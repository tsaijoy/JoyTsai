```python
import pandas as pd
import random

#一種資料結構 #有屬性 分類細節
food_calories = {
    "grains": {
        "Rice": 200,
        "Noodles": 220,
        "Oatmeal": 150,
        "Sweet Potato": 90,
        "Corn": 100,
        "Whole Wheat Bread": 120
    },
    "proteins": {
        "Egg": 70,
        "Tofu": 80,
        "Chicken Breast": 165,
        "Fish": 140,
        "Shrimp": 90,
        "Pork": 210,
        "Beans": 130

}}
```

```python

def generate_daily_meal():
    breakfast = {  #隨機在各種屬性取樣放到各餐裡面 #一層一層執行
        "Grains": random.choice(list(food_calories["grains"].keys())),
        "Protein": random.choice(list(food_calories["proteins"].keys())),
        "Dairy": random.choice(list(food_calories["dairy"].keys())),
        "Fruit": random.choice(list(food_calories["fruits"].keys()))
    }
    lunch = {
        "Grains": random.choice(list(food_calories["grains"].keys())),
        "Protein": random.choice(list(food_calories["proteins"].keys())),
        "Vegetables": random.choice(list(food_calories["vegetables"].keys())),
        "Fruit": random.choice(list(food_calories["fruits"].keys())),
        "Oil": random.choice(list(food_calories["oils"].keys()))
    }
    dinner = {
        "Grains": random.choice(list(food_calories["grains"].keys())),
        "Protein": random.choice(list(food_calories["proteins"].keys())),
        "Vegetables": random.choice(list(food_calories["vegetables"].keys())),
        "Fruit": random.choice(list(food_calories["fruits"].keys())),
        "Oil": random.choice(list(food_calories["oils"].keys()))
    }

    # Calculate calories for each meal
    breakfast_calories = food_calories["grains"][breakfast["Grains"]] + food_calories["proteins"][breakfast["Protein"]] + food_calories["dairy"][breakfast["Dairy"]] + food_calories["fruits"][breakfast["Fruit"]]
    lunch_calories = food_calories["grains"][lunch["Grains"]] + food_calories["proteins"][lunch["Protein"]] + food_calories["vegetables"][lunch["Vegetables"]] + food_calories["fruits"][lunch["Fruit"]] + food_calories["oils"][lunch["Oil"]]
    dinner_calories = food_calories["grains"][dinner["Grains"]] + food_calories["proteins"][dinner["Protein"]] + food_calories["vegetables"][dinner["Vegetables"]] + food_calories["fruits"][dinner["Fruit"]] + food_calories["oils"][dinner["Oil"]]

    total_daily_calories = breakfast_calories + lunch_calories + dinner_calories

    return breakfast, lunch, dinner, total_daily_calories
```
